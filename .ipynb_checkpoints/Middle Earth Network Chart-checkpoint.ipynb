```python
#Import all the packages that are needed
import networkx as ns
import json
import os
import matplotlib as mpl
```

```python
#This function takes the identifier code, and returns the name of the person
def lookup_ident(file,idCode):
    ident=' '
    for person in file['identifier']:
        if idCode==person['identCode']:
            ident= person['name']
    return ident
```

```python
#This function produces a list of tuples of the names of the caller and the recipient
def extract_calls(file):
    call_list=[]
    for call in file['callHistory']:
        caller=lookup_ident(file,call['caller'])
        recipient=lookup_ident(file,call['receiver'])
        combined=(caller,recipient)
        call_list.append(combined)
    return call_list
```

```python
#This function extracts the calls for all "json" files in a directory
def extract_dir(folder):   
    all_calls=[]
    for file in os.listdir(folder):
        if "json" in file:
            all_calls.append(extract_calls(json.load(open(file))))
    #Flatten into one list, rather than a list of lists
    all_calls=[call for sublist in all_calls for call in sublist]
    return all_calls      
```

```python
#Put all calls into one variable called "call_network"
call_network=extract_dir("C:/Users/user/github/Learning Python, JSON and Network Charts/")
```

```python
#Use networkx package (and also matplotlib  from inside that), to produce and visualise a network chart
G=nx.Graph()
G.add_edges_from(call_network)
nx.draw(G,with_labels=True,node_size=200,node_color="Orange")
```
